```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../datasets/wine_quality/winequality-red.csv", sep = ";")
```

```python
data.head()
```

```python
correlation = data.corr()
data.corr()
```

```python
mask = np.triu(np.ones_like(correlation, dtype=bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(500, 20, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation, mask=mask, cmap=cmap, vmax=.3, center=0,square=True, linewidths=.5)
```

```python
data.describe()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
```

```python
x = data.drop("quality",axis=1)
x
```

```python
y = data[["quality"]]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split (x ,y,test_size=0.2,shuffle=False)
```

```python
X_train
```

```python
regresion_lineal = LinearRegression()
```

```python
regresion_lineal.fit(X_train,y_train)
```

```python
regresion_lineal.coef_
```

```python
w_list = regresion_lineal.coef_[0]
columns_names = X_train.columns.tolist()
```

```python
diccionario = dict(zip(columns_names, w_list))
df = pd.DataFrame(diccionario, index=[0])
df
```

```python
regresion_lineal.intercept_
```

```python
prediccion_entrenamiento = regresion_lineal.predict(X_test)
print(prediccion_entrenamiento)
```

```python
score_train = regresion_lineal.score(X_train, y_train)
```

```python
score_test = regresion_lineal.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score
r2_train = r2_score(y_train, regresion_lineal.predict(X_train))
r2_test = r2_score(y_test, regresion_lineal.predict(X_test))
```

```python
print("Puntaje (Score) en entrenamiento:", score_train)
print("Puntaje (Score) en prueba:", score_test)
print("Coeficiente de determinacin R^2 en entrenamiento:", r2_train)
print("Coeficiente de determinacin R^2 en prueba:", r2_test)
```
